## Concept


Here follow my implementation of the TomatoClustering clustering algorithm, presented by Steve Oudot in its book "Persistence Theory: From Quiver Representations to Data Analysis". It mainly consists in exploiting the strength of topological data analysis for clustering. A few examples similar to the ones provided by scikit-learn are given below. The idea behind the algorithm is the following: Given a two-dimensional dataset, the sublevels filtration of the simplextree build over the density estimate of the the data gives, by reconstruction of the visualize of the connected components, the hierarchical relation between the data points. By sorting it, we are able to retrace the density extrema, and thus the centroids. Visualized persistence, as presented below, gives a good indication over the number of centroids. Once determined, the construction of the distinct sets through UnionFind becomes then possible.

## Examples


```python
from topologyx.clustering import ClusterGenerator, ClusterStructure, TomatoClustering
```

```python
generator = ClusterGenerator(structure=ClusterStructure.ANISOTROPY, randomize=20)
tomato = TomatoClustering(*generator.generate())
_ = tomato.estimate_clusters(visualize=True)
_ = tomato.fit_predict(n_clusters=3, visualize=True)
```

```python
generator = ClusterGenerator(structure=ClusterStructure.MOONS, randomize=18)
tomato = TomatoClustering(*generator.generate())
_ = tomato.estimate_clusters(visualize=True)
_ = tomato.fit_predict(n_clusters=2, visualize=True)
```

```python
generator = ClusterGenerator(structure=ClusterStructure.BLOBS, randomize=28)
tomato = TomatoClustering(*generator.generate())
_ = tomato.estimate_clusters(visualize=True)
_ = tomato.fit_predict(n_clusters=2, visualize=True)
```

```python
generator = ClusterGenerator(structure=ClusterStructure.RANDOM, randomize=30)
tomato = TomatoClustering(*generator.generate())
_ = tomato.estimate_clusters(n_neighbors=10, visualize=True)
_ = tomato.fit_predict(n_clusters=6, tau=0.001, visualize=True)
```

```python
generator = ClusterGenerator(structure=ClusterStructure.CIRCLES, randomize=60)
tomato = TomatoClustering(*generator.generate())
_ = tomato.estimate_clusters(n_neighbors=25, visualize=True)
_ = tomato.fit_predict(n_clusters=2, tau=1.0, visualize=True)
```
